```python
import pandas as pd
import numpy as np
from string import ascii_uppercase as alph
import pickle
```

```python
all_tables=pd.read_html("https://web.archive.org/web/20221115040351/https://en.wikipedia.org/wiki/2022_FIFA_World_Cup")
```

```python
#We manually look for the position of the tables
all_tables[12]
all_tables[61]
```

```python
#the tables are found from index 12. and jump 7 places
#we change the name of column one because they don't match.
#We remove the column 'Qualification', because we don't need it
#store all the tables in a dictionary

dict_tables={}
for letter,i in zip(alph, range(12,68,7)):
    df=all_tables[i]
    df.rename(columns={df.columns[1]:"Team"},inplace=True)
    df.pop("Qualification")
    dict_tables[f'Group {letter}']=df 
```

```python
#we save the dictionary in a binary file

with open ('dict_tables','wb')as output:
    pickle.dump(dict_tables,output)
    
```
